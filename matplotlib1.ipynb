```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.title("Graph of x and y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.scatter(x,y)
plt.title("Graph of x and y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.hist(x,y,color="pink",width=1)
plt.title("Graph of x and y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
marks = [40,30,20,10]
subjects = ["Math","Science","English","Computer"]
plt.pie(marks,labels=subjects,autopct="%1.1f%%")
plt.show
```

```python
import matplotlib.pyplot as plt
marks = [40,30,20,10]
subjects = ["Math","Science","English","Computer"]
plt.pie(marks,labels=subjects,autopct="%1.1f%%",explode=[0,0,3,0])
plt.show
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.bar(x,y,color="pink")
plt.title("Graph of x and y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.barh(x,y,color="pink")
plt.title("Graph of x and y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y1 = [2,4,6,8]
y2 = [1,3,5,7]
plt.plot(x,y1,label="Product A")
plt.plot(x,y2,label="Product B")
plt.legend()
plt.show()
```
